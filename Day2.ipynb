```python
checksum=0
rows = []
for r in inputDay2.splitlines():
    row = list(map(int,r.split('\t')))
    rows.append(row)
    theMin = min(row)
    theMax = max(row)
    checksum = checksum + theMax - theMin
checksum
```

```python
import itertools
sum = 0
for row in rows:
    for a, b in itertools.combinations(row, 2):
        if a%b == 0:
            sum = sum + a/b
        elif b%a == 0:
            sum = sum + b/a
sum
```
