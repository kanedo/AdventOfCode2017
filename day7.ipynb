```python
class Node:
    def __init__(self, name, weight, parent, children):
        self.name = name
        self.parent = parent
        self.children = children
        self.weight = int(weight)
    def setParent(self, parent):
        self.parent = parent
        
    def getParent(self):
        return self.parent
        
    def setChildren(self, children):
        self.children = children
    def getChildren(self):
        return self.children
        
    def getName(self):
        return self.name
    def getWeight(self):
        return self.weight
    
```
